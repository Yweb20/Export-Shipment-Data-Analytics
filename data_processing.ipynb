```python
df['SB_date'] = pd.to_datetime(df['SB_date'], errors='coerce') #Converting SB_date to proper datetime.
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df['EGM_Flag'] = df['EGM_no'].apply(
    lambda x: 'Available' if str(x) != 'Not_Available' else 'Not Available'
)
```

```python
df['ROSL_Flag'] = df['ROSL'].apply(
    lambda x: 'Available' if str(x) == 'YES' else 'Not Available'
)
```

```python
df[['ROSL', 'ROSL_Flag']].head()
```

```python
df['GST_Flag'] = df['GST_Compliance_Status'].apply(
    lambda x: 'Compliant' if 'Compliant' in str(x) else 'Non-Compliant'
)
```

```python
df[['GST_Compliance_Status', 'GST_Flag']].head()
```

```python
df['Shipment_Age_Days'] = (pd.Timestamp.today() - df['SB_date']).dt.days
```

```python
port_summary = df.groupby('Port_Code').size().reset_index(name='Total_Shipments')       #Shipments per Port
port_summary 
```

```python
status_summary = df.groupby('Current_status').size().reset_index(name='Count')    #Status wise Shipment Count
status_summary
```

```python
gst_summary = df.groupby('GST_Flag').size().reset_index(name='Shipments')   #GST Compliance summary
gst_summary
```

```python
df.to_csv("Cleaned_Final_Shipment_Data.csv", index=False)       #Saving Final Outputs
port_summary.to_csv("Port_Summary.csv", index=False)
status_summary.to_csv("Status_Summary.csv", index=False) 
gst_summary.to_csv("GST_Summary.csv", index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get("https://www.w3schools.com/html/html_tables.asp")

time.sleep(3)  # allow page to load

rows = driver.find_elements(By.XPATH, '//*[@id="customers"]/tbody/tr')[1:]

data = []

for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    data.append([
        cols[0].text,
        cols[1].text,
        cols[2].text
    ])

driver.quit()

df_web = pd.DataFrame(data, columns=["Company", "Contact", "Country"])

print(df_web.head())
print("Rows scraped:", len(df_web))

df_web.to_csv("selenium_extracted_data.csv", index=False)
```
